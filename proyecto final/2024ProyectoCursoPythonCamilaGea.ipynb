```python
#  Importar Libreria Pandas
import pandas as pd

#  Leer un archivo de Excel
convertir=pd.read_excel('CabildoPandas.xlsx')

#  Convertir a .csv
convertir.to_csv('CabildoPandas.csv',index=None, header=True)

#  Leer con Pandas el dataframe
df=pd.read_csv('CabildoPandas.csv')

#  Mostrar dataframe
df
```

```python
#  Mostrar información básica del dataframe
df.describe()
```

```python
#  Nuevo dataframe
newdf=df.copy()
```

```python
#  Convertir la columna de fecha en tipo de dato fecha
newdf['Unnamed: 0']=pd.to_datetime(newdf['Unnamed: 0'])
```

```python
#  Importar Libreria Matplotlib para poder graficar
import matplotlib.pyplot as plt

#  Grafico lineal

#  Editar el tamaño de la figura
plt.figure(figsize=(15,6))

# Datos para graficar
x= newdf['Unnamed: 0']
y= df['PB- Secretaría Administrativa']

x2= newdf['Unnamed: 0']
y2= df['PB- Oficina Director']

x3= newdf['Unnamed: 0']
y3= df['PB- Boletería y Seguridad']


#  Configuracion de las curvas
plt.plot(x,y,color='r',label='Secretaria Administrativa')
plt.plot(x2,y2,color='b',label='Oficina Director')
plt.plot(x3,y3,color='c',label='Boleteria y Seguidad')

#  Configuracion del gráfico
plt.xlabel('Fecha')
plt.ylabel('Temperatura [°C]')
plt.title('Temperatura Cabildo Verano')
plt.legend()

#  Configuracion de los ejes
plt.yticks([18,20,22,24,26,28,30,32,34,36])
plt.xticks(rotation=30)

#  Mostrar el gráfico
plt.show


```

```python
# Grafico de dispersión

x3= df['Unnamed: 4']
y3= df['PB- Boletería y Seguridad']

plt.scatter(y3,x3,color='c')
plt.show()
```

```python
#  Graficos lineales separados

fig, ax=plt.subplots(1,2,sharey=True)

ax[0].plot(x,y,color='g', label='Secretaria Administrativa')
ax[0].legend()

ax[1].plot(x,y,color='c',label='Oficina Director')
ax[1].legend()

plt.show()
```
